```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\UC\\Desktop\\covid-19_data_Nigeria.csv")
data.head()
```

```python
data.describe()
```

Average age of confirmed cases

```python
sns.set()
sns.distplot(data['Avg_age'], kde=False,  color='b'  )

plt.show()
```

Majority of confirmed cases is between the ages of 65-75 years.

The majority of confirmed death is between the ages of 75 and 80 years. According to the graph below

```python
sns.set()
sns.distplot(data['Avg_age_of_death'], kde=False,  color='b'  )

plt.show()
```

```python
sns.relplot(x='month_since_first_confirmed', y='Confirmed', kind='line', data=data, hue='Region', ci=None)
```

Comparing confirmed deaths and recovered

```python
sns.relplot(x='Confirmed', y='Deaths', kind='scatter', data=data)
plt.show()
```

From the graph above, we can confirm that the more confiirmed cases, the more death.

```python
sns.relplot(x='Confirmed', y='Deaths', kind='scatter', data=data, hue='Region')
plt.show()
```

```python
sns.jointplot(x='Confirmed', y='Deaths', kind='scatter', data=data, hue='Region')
plt.show()
```

```python
sns.jointplot(x='Confirmed', y='Deaths', kind='reg', data=data)
plt.show()
```

```python
sns.jointplot(x='Confirmed', y='Deaths', kind='kde', data=data)
plt.show()
```
